```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
```

```python
# img_path = 'zebra.jpg'
img_path = 'linjal.jpg'
# img_path = 'tennisboll.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)
```

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

## importera moduler från tensonflow
```

```python
model = ResNet50(weights='imagenet') ## Avänds för bildklacifering
```

```python
## Välj en bild
# img_path = 'Sudd.jpg'
# img_path = 'Penna.jpg'
# img_path = "fotboll.png"
img_path = "panda.jpg"
img = image.load_img(img_path, target_size=(224, 224))
img  ##Printer ut bilden
```

```python
x = image.img_to_array(img)   ##Omvandndling till numpyarray
x.shape  ### Skriver ut diminsioner
```

```python
x = np.expand_dims(x, axis=0)  ## UTökar diminsioner med index 0
x.shape  ## Visar formen
```

```python
x = preprocess_input(x) ## Förbreda för att passa resnet50 modeller för förutsägelse
```

```python
preds = model.predict(x)  ## Generera förutsägelse
print("Predicted:", decode_predictions(preds, top=3)[0])  ##Printar ut förutsägelserna
```
